```python
from sklearn.datasets import make_classification
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

# Generate synthetic data
X, y = make_classification(n_samples=20, n_features=4, n_classes=2)

# Function to get K-fold indices
def kfold_indices(data, k):
    fold_size = len(data) // k
    indices = np.arange(len(data))
    folds = []
    for i in range(k):
        test_indices = indices[i * fold_size: (i + 1) * fold_size]
        train_indices = np.concatenate([indices[:i * fold_size], indices[(i + 1) * fold_size:]])
        folds.append((train_indices, test_indices))
    return folds

# 1) Perform K-Fold Cross-validation with number of samples = 20 and number of splits = 5.
k_1 = 5
fold_indices_1 = kfold_indices(X, k_1)
model_1 = DecisionTreeClassifier()
scores_1 = []

for train_indices, test_indices in fold_indices_1:
    X_train, y_train = X[train_indices], y[train_indices]
    X_test, y_test = X[test_indices], y[test_indices]
    model_1.fit(X_train, y_train)
    y_pred = model_1.predict(X_test)
    fold_score = accuracy_score(y_test, y_pred)
    scores_1.append(fold_score)

mean_accuracy_1 = np.mean(scores_1)
print("K-Fold Cross-Validation Scores (K=5):", scores_1)
print("Mean Accuracy (K=5):", mean_accuracy_1)

# 2) Perform K-Fold Cross-validation with number of samples = 20 and number of splits = 10.
k_2 = 10
fold_indices_2 = kfold_indices(X, k_2)
model_2 = DecisionTreeClassifier()
scores_2 = []

for train_indices, test_indices in fold_indices_2:
    X_train, y_train = X[train_indices], y[train_indices]
    X_test, y_test = X[test_indices], y[test_indices]
    model_2.fit(X_train, y_train)
    y_pred = model_2.predict(X_test)
    fold_score = accuracy_score(y_test, y_pred)
    scores_2.append(fold_score)

mean_accuracy_2 = np.mean(scores_2)
print("\nK-Fold Cross-Validation Scores (K=10):", scores_2)
print("Mean Accuracy (K=10):", mean_accuracy_2)
```
