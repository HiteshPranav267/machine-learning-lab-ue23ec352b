```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
df = pd.DataFrame(iris.data)
label = pd.DataFrame(iris.target)

# Step 1: Test = 20%
X_train_full, X_test, y_train_full, y_test = train_test_split(
    df, label, test_size=0.20, shuffle=True
)

# Step 2: Validation from remaining
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.1875, shuffle=True
)

print("Training:", X_train.shape[0])
print("Validation:", X_val.shape[0])
print("Testing:", X_test.shape[0])
```

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
df = pd.DataFrame(iris.data)
label = pd.DataFrame(iris.target)

# Train = 75%
X_train, _, y_train, _ = train_test_split(
    df, label, test_size=0.25, shuffle=True
)

# Validation = 15%
_, X_val, _, y_val = train_test_split(
    df, label, test_size=0.15, shuffle=True
)

# Test = 15%
_, X_test, _, y_test = train_test_split(
    df, label, test_size=0.15, shuffle=True
)

print("Training samples:", X_train.shape[0])
print("Validation samples:", X_val.shape[0])
print("Testing samples:", X_test.shape[0])

print("Total effective samples:",
      X_train.shape[0] + X_val.shape[0] + X_test.shape[0])
```

```python
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split

# Load diabetes dataset
diabetes = datasets.load_diabetes()
df = pd.DataFrame(diabetes.data)
label = pd.DataFrame(diabetes.target)

# Step 1: Test = 25%
X_remain, X_test, y_remain, y_test = train_test_split(
    df, label, test_size=0.25, shuffle=True
)

# Step 2: Train & Validation from remaining (70:30)
X_train, X_val, y_train, y_val = train_test_split(
    X_remain, y_remain, test_size=0.30, shuffle=True
)

# Print sizes
print("Total samples:", df.shape[0])
print("Training samples:", X_train.shape[0])
print("Validation samples:", X_val.shape[0])
print("Testing samples:", X_test.shape[0])
```
