## Iris dataset

```python
from sklearn import datasets
import pandas as pd
```

```python
print("THIS IS A PROGRAM TO ACCESS DATASET")
iris = datasets.load_iris()
```

```python
print('DESCRIPTION OF DATASET')
print(iris.DESCR)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['target'] = pd.Series(iris.target)
```

```python
df['target_names'] = df['target'].apply(lambda y:iris.target_names[y])
```

```python
print('Displaying First 5 samples')
print(df.head(5))
```

```python
print('Displaying random 5 samples')
print(df.sample(5))
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.3)
```

```python
print("Total number of samples in Iris dataset = ", df.shape[0])
print("Total number of training samples in Iris dataset = ", df_train.shape[0])
print("Total number of testing samples in Iris dataset = ", df_test.shape[0])
```

```python
print('First 5 samples of training set: ')
print(df_train.head(5))
```

```python
print('First 5 samples of testing set: ')
print(df_test.head(5))
```
